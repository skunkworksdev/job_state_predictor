```python
# The above cell would create a JSON file
# Directly calling pandas read_json doesnt seem to work (some length mismatch error)
# json.load on the File object throws extra data error.
# Parse the JSON line by line
import json
jobs = []


with open('1711670203-anon_jobs_2019-12.json', 'r') as f:
#with open('1711670203-anon_jobs_2020-04.json', 'r') as f:
#with open('1711670203-anon_jobs_2020-08.json', 'r') as f:
  for job in f:
    jobs.append(json.loads(job))

jobs[:5]
```

```python
columns = ['job_id', 'user', 'account', 'script', 'qos', 'array_pos',
           'partition', 'submit_time', 'start_time', 'queue_wait', 'end_time',
           'processors_req', 'processors_used', 'nodes_req', 'nodes_used',
           'nodelist', 'wallclock_req', 'wallclock_used', 'cpu_used',
           'std_power', 'avg_power',  'state']

# Create an array of jobs.
# Each job is a list containing the above fields in the same order
jobs_array = []
for job in jobs:
  job_details = []
  for column in columns:
    if column in job.keys():
      job_details.append(job[column])
    else:
      job_details.append(None)
  jobs_array.append(job_details)

jobs_array[:5]
```

```python
import pandas as pd


jobs_df = pd.DataFrame(data=jobs_array, columns=columns)
jobs_df.head()
```

```python
# user, account and script columns are anonymized.
# Find the value counts for these columns.
# If all the values in these columns are distinct, they are useless

for column in ['user', 'account', 'script']:
  print("\nValue count for ",column)
  print(jobs_df[column].value_counts())
```

```python
# It is evident from above that script is not a useful column (41658 distinct values).
# Job ID is also useless, since its unique for every job.
# Hence, drop these two columns

jobs_df.drop(columns=['job_id', 'script'], inplace=True)
jobs_df.shape
```

```python
# Check if nodes requested = nodes used.
# If shape for the below code is 0,19, above condition will hold true.

jobs_df[jobs_df['nodes_req'] != jobs_df['nodes_used']].shape
```

```python
# Above condition does not hold true.
# This is because users might have specified a greater number of tasks than the available number of CPUs on the node.

# First, find the jobs where only one node has been used.
# Second, find the value counts for the processors_used column.
# If there is only one value, it mostly means that the each node has the same number of processors.

one_node_jobs = jobs_df[jobs_df['nodes_used'] == 1]
one_node_jobs['processors_used'].value_counts()
```

```python
# There are very few jobs that are allocated a node with a different number of processors. Let's analyze this record.

one_node_jobs[one_node_jobs['processors_used'] == 36864]
```

```python
# We need to check the length of the nodelist.

len(one_node_jobs[one_node_jobs['processors_used'] == 36864]['nodelist'].iloc[0])
```

```python
# In above case, the number of processors = 1024 * 36.
# There's a chance that somehow, these jobs got nodes_used incorrectly as 1.
# Let's verify if nodes_used is always equal to the length of nodelist.

nodelist_length = jobs_df['nodelist'].apply(lambda x: len(x))
jobs_df[jobs_df['nodes_used'] != nodelist_length].shape
```

```python
# As expected, very few jobs have got it wrong.
# We can correct those job's values.
# After changing, check whether the change is done.

incorrect_job_index = jobs_df[jobs_df['processors_used'] == 36864].index
jobs_df.loc[incorrect_job_index, ['nodes_used']] = 1024
nodelist_length = jobs_df['nodelist'].apply(lambda x: len(x))
jobs_df[jobs_df['nodes_used'] != nodelist_length].shape
```

```python
# Now, check whether processors_used = 36 * nodes_used

calculated_processors_used = jobs_df['nodes_used'].apply(lambda x: 36 * x)
jobs_df[jobs_df['processors_used'] != calculated_processors_used].shape
```

```python
# Above condition holds true. We can drop processors_used.
# Check value_counts of array_pos.

jobs_df['array_pos'].value_counts()
```

```python
# We can't drop array_pos as it contains a decent number of values.
# Let's check if nodes_req is the same as processors_req

jobs_df[jobs_df['processors_req'] != jobs_df['nodes_req']].shape
```

```python
# Since above condition does not hold true, analyze.

df_subset = jobs_df[jobs_df['processors_req'] != jobs_df['nodes_req']]
df_subset.head()
```

```python
# It seems that nodes_req != processors_req for the same reason for which nodes_req != nodes_used; User options
# Hence we can't drop processors_req and nodes_req.
# One pattern is visible: if processors_req <= 36, nodes_req = 1. If processors_req > 36, nodes_req = 2.
# Hence, if processors_req > K * 36, node_req = K + 1.
# Or, nodes_req = ceil(processors_req / 36).
# Check if the above condition holds true
import math

calculated_nodes_req = df_subset['processors_req'].apply(lambda x: math.ceil(x / 36))
df_subset[df_subset['nodes_req'] != calculated_nodes_req].shape
```

```python
# Above condition doesnt hold. Analyze.

df_subset1 = df_subset[df_subset['nodes_req'] != calculated_nodes_req]
df_subset1.head()
```

```python
# User options are the reason. Hence, we can drop neither nodes_req nor processors_req.
# Now's let's analyze wallclock_used and cpu_used.
# Before that, convert wallclock_req, wallclock_used and cpu_used into seconds.
# Each string of these fields is as follows: P<Num_days>T<Num_hours><Num_mins><Num_secs>.

import re

def convert_to_seconds(string):
  obj = re.findall('\d+', string)
  seconds = int(obj[0]) * 86400 +  int(obj[1]) * 3600 + int(obj[2]) * 60 + int(obj[3])
  return seconds

for i in ["wallclock_req", "wallclock_used", "cpu_used"]:
  jobs_df[i] = jobs_df[i].apply(convert_to_seconds)

jobs_df.head()
```

```python
# Check if cpu_used = wallclock_used * 36 * nodes_used

calculated_cpu_used = jobs_df['wallclock_used'] * 36 * jobs_df['nodes_used']
jobs_df[jobs_df['cpu_used'] != calculated_cpu_used].shape
```

```python
# Above condition holds true. We can drop cpu_used.
# Now, check if wallclock_used <= wallclock_req + 10s. This is because Slurm tries to stop the job if it exceeds wallclock_req.

jobs_df[jobs_df['wallclock_used'] > jobs_df['wallclock_req'] + 10].shape
```

```python
# Above condition does not hold true. Analyze.

df_subset2 = jobs_df[jobs_df['wallclock_used'] > jobs_df['wallclock_req'] + 10]
df_subset2.head()
```

```python
# It seems that jobs runs for a variable number of seconds after the time limit before stopping.
# Instead of 10s, let's find the max number of seconds that the job can run after the time limit.
# In Slurm methodology, Time Limit is nothing but wallclock_req.

#jobs_df[jobs_df['wallclock_used'] > jobs_df['wallclock_req'] + 80].shape
#jobs_df[jobs_df['wallclock_used'] > jobs_df['wallclock_req'] + 150].shape
jobs_df[jobs_df['wallclock_used'] > jobs_df['wallclock_req'] + 50000].shape
```

```python
# After quite a lot of experimentation, it seems that all the jobs stop after 80s.
# Another dataset (2020-08), saw this time duration being 50000.
# Jobs that exceed time limit have their states marked as TIMEOUT.
# Let's verify if the above condition holds true.

df_subset3 = jobs_df[jobs_df['wallclock_used'] > jobs_df['wallclock_req']]
df_subset3[df_subset3['state'] != "TIMEOUT"].shape
```

```python
# Above condition does not hold for only two records. Analyze.

weird_jobs = df_subset3[df_subset3['state'] != "TIMEOUT"]
weird_jobs
```

```python
# The above jobs have failed. They are a special case of TIMEOUT since after exceeding the time limit, they failed, while the rest were stopped by Slurm.
# Sometimes, jobs can complete or be cancelled after timeout. (2020-08).
# We drop wallclock_used not only because it can be obtained only after the end of a job, but also that in most cases it is less than wallclock_req.
# std_power and avg_power are measured during the execution of a job.
# The measurements are taken at periodic intervals in the duration when a job runs.
# Hence, from the point of submission time or even allocation time, these metrics cannot be obtained.
# Let's drop all the columns that were identified for removal so far.

jobs_df.drop(columns=['processors_used', 'cpu_used', 'wallclock_used', 'std_power', 'avg_power'], inplace=True)
jobs_df.head()
```

```python
# We can additionally drop end_time since its not required by JSP.
# queue_wait is the difference between start_time and submit_time; we can drop queue_wait too.

jobs_df.drop(columns=['end_time', 'queue_wait'], inplace=True)
jobs_df.head()
```
